```python
import pandas as pd
df=pd.read_csv('Final_project_dataset')
df.head()
```

```python
inputs=df.drop(['loan_grade'],axis=1)
```

```python
target=df['loan_grade']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_person_home_ownership=LabelEncoder()
le_loan_intent=LabelEncoder()
le_cb_person_default_on_file=LabelEncoder()
```

```python
inputs['home_ownership']=le_person_home_ownership.fit_transform(inputs['person_home_ownership'])
inputs['loan_intent']=le_person_home_ownership.fit_transform(inputs['loan_intent'])
inputs['person_default_on_file']=le_person_home_ownership.fit_transform(inputs['cb_person_default_on_file'])
```

```python
inputs.head()
```

```python
inputs_n=inputs.drop(['Unnamed: 0', 'person_home_ownership','loan_intent','cb_person_default_on_file'],axis=1)
```

```python
inputs_n.head()
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold, cross_val_score

```

```python

X, y = datasets.load_iris(return_X_y=True)

clf = DecisionTreeClassifier(random_state=42)

k_folds = KFold(n_splits = 5)

scores = cross_val_score(clf, X, y, cv = k_folds)

print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
print("Number of CV Scores used in Average: ", len(scores))
```

```python
model.fit(inputs_n.values,target)
```

```python
model.score(inputs_n.values,target)
```
